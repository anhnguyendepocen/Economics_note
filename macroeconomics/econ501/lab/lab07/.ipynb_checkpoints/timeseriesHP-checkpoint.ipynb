DGP 2 (non-stationary data):

````python
%%stata 

gen x = 0 if t==1

gen e1 = rnormal() 

replace x=L.x + e1 if t>1 

gen u = 0 if t==1 

gen e2 = rnormal()

replace u =L.u + e2 if t>1 

gen y = u


*/* White noise error */ x is a random walk
*/* White noise error */ u is a random walk
*Initial value of x is 0 */ */
*Initial value of u is 0 */ */
````

````python
%%stata --graph
* random walk
tsline y
````

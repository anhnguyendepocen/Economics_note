



A series with drift can be modeled as yt=c+ϕyt−1+ϵt where cc is the drift(constant), 


and ϕ=1


A series with trend can be modeled as yt=c+δt+ϕyt−1+ϵt where cc is the drift(constant), δt is the deteministic time trend and ϕ=1

the drift term in your equation with ϕ=1ϕ=1 generates a deterministic linear trend in the observed series, while a deterministic trend turns into an exponential pattern in ytyt.

To see what I mean, you could simulate and plot some series with the R software as shown below.

Simulate a random walk:

```r
n <- 150
eps <- rnorm(n)
x0 <- rep(0, n)
for (i in seq.int(2, n))
  x0[i] <- x0[i-1] + eps[i]
plot(ts(x0))
```

Simulate a random walk with drift:

```r
drift <- 2
x1 <- rep(0, n)
for (i in seq.int(2, n))
  x1[i] <- drift + x1[i-1] + eps[i]
plot(ts(x1))
```

Simulate a random walk with a deterministic trend:

```r
trend <- seq_len(n)
x2 <- rep(0, n)
for (i in seq.int(2, n))
  x2[i] <- trend[i] + x2[i-1] + eps[i]
plot(ts(x2))
```

###  8. Euler's formula

https://www.youtube.com/watch?v=VG9mI7Kw89I

For students in science and engineering, you will often use exponential notation to abbreviate the polar form $\cos (\theta )+i\sin (\theta )$  of a complex number of modulus 1. This equation is known as Euler's Formula :

$$e^{i\theta }=\cos (\theta )+i\sin (\theta ).$$

Let's justify Euler's formula by applying what we've learned about differential equations thus far.

The function $e^ t$ is defined to be the solution of the initial value problem , . More generally, the chain rule implies the

$$\textrm{For any constant }\,  a, \,  \,  e^{at} \,  \textrm{ is the solution of } \,  \,  \dot x=ax, \quad x(0)=1.$$

In addition to solving this initial value problem, for Euler's formula to be a good definition, we want  to satisfy exponential properties we know from Calculus. That is to justify Euler's formula, we expect to have

Exponential law:  $$\displaystyle e^{it_1}e^{it_2} = e^{i(t_1+t_2)}$$

Initial value problem: $$\displaystyle \frac{d}{dt} e^{it} = ie^{it}$$,  $$e^{i0} = 1$$.

Taylor's formula: $$\displaystyle e^{it} = \sum _{n=0}^{\infty } \frac{(it)^ n}{n!}.$$

The next videos justify the first two properties of the exponential $e^{it}$. You may enjoy verifying Taylor's formula yourself.

$$\displaystyle \frac{d}{dt} e^{i t} = ie^{it}.$$

$$\displaystyle  \cos (t_1 + t_2) + i \sin (t_1+t_2)$$

### A first order homogeneous differential equation

```python
#https://www.youtube.com/watch?v=4X0SGGrXDiI

a, t, y0 = sym.symbols("a t y0")
y = sym.Function("y")
```

```python
ode=sym.Eq(y(t).diff(t), a*y(t))
ode
```

- y: money in bank
- a: interest rate, for example 6%
- t: time year


```python
sym.classify_ode(ode)
```

```python
ode_sol=sym.dsolve(ode, y(t), hint="separable")
ode_sol
```

```python
ode_sol.free_symbols
```

```python
ics = {y(0): y0}
```

```python
def apply_ics(sol, ics, x, known_params):
    """
    Apply the initial conditions (ics), given as a dictionary on
    the form ics = {y(0): y0: y(x).diff(x).subs(x, 0): yp0, ...}
    to the solution of the ODE with indepdendent variable x.
    The undetermined integration constants C1, C2, ... are extracted
    from the free symbols of the ODE solution, excluding symbols in
    the known_params list.
    """
    free_params = sol.free_symbols - set(known_params)
    eqs = [(sol.lhs.diff(x, n) - sol.rhs.diff(x, n)).subs(x, 0).subs(ics)
           for n in range(len(ics))]
    sol_params = sym.solve(eqs, free_params)
    return sol.subs(sol_params)
```

```python
apply_ics(ode_sol, ics, t, [a])
```

### Second Order Differential Equation

#### Define the differential equation as a sym.Eq()

```python
x,t, b,c,d, n, C0, y0 = sym.symbols("x t  b c d n C0 y0")
f = sym.Function("f")
```

```python
diffeq = sym.Eq(f(t).diff(t, t) - 2*f(t).diff(t) + f(t), sym.sin(t))
diffeq
```

#### Solve differential equation

```python
soln = sym.dsolve(diffeq,f(t))
soln
```

## Properties of Quasi-concave and Quasi-convex functions

1.        A linear function is both quasi-convex and quasi-concave.

2)       All concave (convex) functions (strict or non strict) are quasi-concave (quasi-   convex). But the opposite is not true.

3)       If f(x) quasi convex (strict or non-strict) then –f(x) is quasi-concave

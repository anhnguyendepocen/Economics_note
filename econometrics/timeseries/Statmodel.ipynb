Scikit-learn focus on prediction of time series by machine learning algorithm. It doesn't contain traditional time series analysis such as ARMA, VAR, and relevant statictical tests. But you can find these in Statsmodels package.

```python
dta = sm.datasets.sunspots.load_pandas().data  #use the dataset sunspot
print(sm.datasets.sunspots.NOTE)
```

```python
dta.index = pd.Index(sm.tsa.datetools.dates_from_range('1700', '2008')) #set time series
del dta["YEAR"]
```

```python
dta.plot(figsize=(12,8))
```

```python
fig = plt.figure(figsize=(12,8))
ax1 = fig.add_subplot(211)
fig = sm.graphics.tsa.plot_acf(dta.values.squeeze(), lags=40, ax=ax1)
ax2 = fig.add_subplot(212)
fig = sm.graphics.tsa.plot_pacf(dta, lags=40, ax=ax2)
```
